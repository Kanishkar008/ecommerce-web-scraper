```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

books = []

for page in range(1, 6):

    url = f"https://books.toscrape.com/catalogue/page-{page}.html"

    try:
        response = requests.get(url)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")

        products = soup.find_all("article", class_="product_pod")

        for product in products:

           
            title_tag = product.find("h3").find("a")
            title = title_tag.get("title", "N/A")

           
            price_tag = product.find("p", class_="price_color")
            price = price_tag.text.strip() if price_tag else "N/A"
            price = price.replace("Â£", "£")

          
            rating_tag = product.find("p", class_="star-rating")
            if rating_tag:
                classes = rating_tag.get("class")
                rating = classes[1] if len(classes) > 1 else "N/A"
            else:
                rating = "N/A"

            
            availability_tag = product.find(
                "p",
                class_="instock availability"
            )

            availability = (
                availability_tag.text.strip()
                if availability_tag
                else "N/A"
            )

            books.append({
                "Title": title,
                "Price": price,
                "Rating": rating,
                "Availability": availability
            })

        print(f"Page {page} scraped successfully.")

    except Exception as e:
        print(f"Error scraping page {page}: {e}")


df = pd.DataFrame(books)


df.to_csv("books_dataset.csv", index=False, encoding="utf-8")


print("\nScraping Completed!")
print(f"Total Products Scraped: {len(df)}")

print("\nFirst 5 Records:")
print(df.head())

print("\nCSV file saved as 'books_dataset.csv'")
```
